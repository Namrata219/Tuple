Tuple Items

Tuple items are ordered, unchangeable, and allow duplicate values.


Tuple items are indexed, the first item has index [0], the second item has index [1] etc.

Ordered
When we say that tuples are ordered, it means that the items have a defined order, and that order will not change.

Unchangeable
Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created.

Allow Duplicates
Since tuples are indexed, they can have items with the same value:

```python
x= tuple()
x= (1,2, 'learning')
print(x)
print(type (x))
```

```python
x1=(1,2,'mango','paris')
print(x1)
```

```python
x1=(1,2,'mango','paris', 'paris')
print(x1)
print(len(x1))
```

Create Tuple With One Item

To create a tuple with only one item, you have to add a comma after the item, otherwise Python will not recognize it as a tuple

```python
x1=(1)
x2=('mango') 
x=('mango',) #for a tuple comma is imp

print(type(x1))
print(type(x2))
print(type(x))
```

```python
#Tuples are unchangeable, meaning that you cannot change, add, or remove items once the tuple is created.
#But there are some workarounds : like converting tupple in to list add,change the item and convert list again into tupple
```

```python
x = ("apple", "banana", "cherry","mango",'peru','pineapple','mustard')
y= list(x)
y.append('orange')
x=tuple(y)
print(x)
```

```python
#2. Add tuple to a tuple. You are allowed to add tuples to tuples, so if you want to add one item,
#(or many), create a new tuple with the item(s), 
#and add it to the existing tuple:
```

```python
x = ("apple", "banana", "cherry","mango",'peru','pineapple','mustard')
y= ('orange',) # adding comma is must else it will consider as string
x +=y
print(x)
```

```python
#Removal 
#Tuples are unchangeable, so you cannot remove items from it, but you can use the same workaround as we used for changing and
#adding tuple items:
```

```python
x = ("apple", "banana", "cherry","mango",'peru','pineapple','mustard')
y= list(x)
y.remove('peru')
x=tuple(y)
print(x)
```

```python
#When we create a tuple, we normally assign values to it. This is called "packing" a tuple:
#in Python, we are also allowed to extract the values back into variables. This is called "unpacking":
```

```python
x = ("apple", "banana", "cherry") #packing
x
```

```python
x = ("apple", "banana", "cherry") #unpacking
(a,b,c)= x
print(a)
print(b)
print(c)
```

```python
#Note: The number of variables must match the number of values in the tuple, 
#if not, you must use an asterisk to collect the remaining values as a list.
```

```python
x = ("apple", "banana", "cherry", "strawberry", "raspberry")

(a, b, *c) = x

print(a)
print(b)
print(c)
```

```python
#If the asterisk is added to another variable name than the last,
#Python will assign values to the variable until the number of values left matches the number of variables left.
```

```python
x = ("apple", "banana", "cherry", "strawberry", "raspberry")

(a, *b, c) = x

print(a)
print(b)
print(c)
```

```python
#Using a While Loop
#You can loop through the list items by using a while loop.

#Use the len() function to determine the length of the tuple, then start at 0 and 
#loop your way through the tuple items by refering to their indexes.

#Remember to increase the index by 1 after each iteration.
```

```python
x = ("apple", "banana", "cherry")
i = 0
while i < len(x):
  print(x[i])
  i = i + 1
```

```python
x = ("apple", "banana", "cherry")
i = 0
while i < len(x):
  print(x)
  i = i + 1
```
